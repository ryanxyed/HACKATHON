# **TRANSFORM**

**step 4 :** processing the data according to the requirements

transforming customers_df

```

#renaming the columns with new column names

customers_df.columns = ['customer_id', 'first_name', 'last_name','phone','city','state','signup_date']
customers_df
```

```

# removing the first row because it is dummy

customers_df.drop(0,axis=0,inplace=True)
customers_df
```

```

#concatenating the 'first_name'+'last_name' to from a new column 'customer_name'

customers_df['customer_name']=customers_df['first_name']+' '+customers_df['last_name']
customers_df
```

```

# removing the 'first_name' and 'last_name' because of 'customer_name'

customers_df.drop(['first_name','last_name'],axis=1,inplace=True)
customers_df
```

```

#function to clean the 'phone' column

def clean_number(number):
  new=''
  for ch in number:
    if ch.isdigit():
      new+=ch
  return new[-10:] # return last 10 digits
```

```

# replace old 'phone' column with new cleaned 'phone' column

customers_df['phone']=customers_df['phone'].apply(clean_number)
customers_df
```

```

# reordering the columns in a desired order

customers_df=customers_df[['customer_id','customer_name','phone','city','state','signup_date']]
customers_df
```

transforming products_df

```
cur.execute('select product_id,product_name,product_category_raw,product_category from products')
products_df=pd.DataFrame(cur.fetchall())
products_df.columns=['product_id','product_name','product_category_raw','list_price']
products_df.drop(0,axis=0,inplace=True)
def clean_category(raw):
  raw=raw.lower()
  if 'elec' in raw:
    raw= 'electronics'
  if '0' in raw:
    raw=raw.replace('0','o')
  if ' ' in raw:
    raw=raw.replace(' ','')
  if raw=='book':
    raw='books'
  if raw=='beauty&care':
    raw='beauty'
  if 'app' in raw:
    raw='apparel'
  raw=raw.title()
  return raw
products_df['product_category']=products_df['product_category_raw'].apply(clean_category)
products_df['product_category_raw']=products_df['product_category']
products_df.drop('product_category',axis=1,inplace=True)
products_df = products_df.rename(columns={'product_category_raw': 'product_category','list_price':'price'})


print(products_df)
```

transforming payments_df

```
cur.execute('select payment_id,order_id,payment_method,amount,payment_date from payments')
payments_df=pd.DataFrame(cur.fetchall())
payments_df.columns=['payment_id','order_id','payment_method','amount','payment_date']
payments_df.drop(0,axis=0,inplace=True)
print(payments_df)
```

transforming date_df

```
cur.execute('select order_id,DAYNAME(order_date),WEEK(order_date),MONTHNAME(order_date),QUARTER(order_date) from orders')
date_df=pd.DataFrame(cur.fetchall())
date_df.columns=['order_id','days','weeks','months','quarters']
date_df.drop(0,axis=0,inplace=True)
print(date_df)
```

transforming order_items_df

```
cur.execute('select order_item_id,order_id,product_id,unit_price,quantity,line_total from order_items')
order_items_df=pd.DataFrame(cur.fetchall())
order_items_df.columns=['order_item_id','order_id','product_id','unit_price','quantity','line_total']
order_items_df.drop(0,axis=0,inplace=True)
print(order_items_df)
```

transforming returns_df

```
cur.execute('select return_id,order_id,returned_date,refund_amount from returns')
returns_df=pd.DataFrame(cur.fetchall())
returns_df.columns=['return_id','order_id','return_date','refund_amount']
returns_df.drop(0,axis=0,inplace=True)
print(returns_df)
```

intialising fact_sales_df

```
cur.execute('select order_id,customer_id,order_date,status,channel from orders')
fact_sales_df=pd.DataFrame(cur.fetchall())
fact_sales_df.columns=['order_id','customer_id','order_date','status','order_value']
fact_sales_df.drop(0,axis=0,inplace=True)
print(fact_sales_df)
```
